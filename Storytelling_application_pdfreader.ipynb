```python
from PyPDF2 import PdfReader
from google.colab import files
uploaded=files.upload()
file_name=next(iter(uploaded))
from gtts import gTTS
from IPython.display import Audio
with open(file_name,'rb')as book:
  reader=PdfReader(book)
  pages=len(reader.pages)
  print('number of pages',pages)
  pgnumber=int(input('Enter the page number you would want to read :'))
  page=reader.pages[pgnumber]
  answer=page.extract_text()
  print('text on page')
  print(answer)

print('The available languages are :English ,French ,Spanish and Portugese')
LANG=input('enter the language you would like to hear your story in :')
if LANG=='English' :
  Lang="en"
elif LANG=='French' :
  Lang='fr'
elif LANG=='Spanish':
  Lang='es'
else:
  Lang='pt'
accents={'English':['UK ','US','Australian '],'French':['Canadian','France'],
         'Portugese':['Portugal ','Brazil '],
         'Spanish':["Spain ",'Mexico ','US']}
print('The different accents available are :',accents)
ACCENT=input('enter the preffered accent of the chosen language :')
if ACCENT=="UK":
  Accent='co.uk'
elif ACCENT=='US':
  Accent='us'
elif ACCENT=='Australian':
  Accent='com.au'
elif ACCENT=='Canandian':
  Accent='ca'
elif ACCENT=='France':
  Accent='fr'
elif ACCENT=='Portugal':
  Accent="pt"
elif ACCENT=='Brazil':
  Accent='com.br'
elif ACCENT=='Spain':
  Accent='es'
else:
  Accent='com.mx'
t=answer
tts=gTTS(text=t,lang=Lang,tld=Accent)
audio_file='output.mp3'
tts.save(audio_file)
Audio(audio_file)
```
